Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib

Matplotlib is open source library in python. While analyzing data, along with statistical inferences it becomes essential to visualize your data, and Matplotlib provides a range of plotting library.

Mtalplotlib is a massive library and has lots of modules, most commonly used -
pyplot, that offer generating different plots.

Types of plots:
1. Bar chart
2. Line chart
3. Histogram
4. Scatter
5. 3d scatter plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = ['a', 'b' , 'c','d','e']
y = np.random.rand(5)
```

```python
plt.Figure(figsize = (5,3))
plt.bar(x,y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Bar chart')
```

```python
z=np.random.rand(5)
plt.scatter(x,y)
plt.title('Scatter plot')
```

```python
h = [1,2,3,3,4,5,5,6,3,2,4]
plt.hist(h)
plt.title('Histogram')
```

```python
plt.plot(y,z)
plt.title('Line chart')
```

```python
l = np.random.rand(25)
m = np.random.rand(25)
n = np.random.rand(25)
fig = plt.figure()                                                    # creating a figure object (a blank canvas where elements can be added)
ax = fig.add_subplot(projection = '3d')
ax.scatter(l,m,n,c='#9327E8',alpha =0.9)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title('3d scatter plot')

plt.show()
```

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))


Note: Also add title, xlabel, and ylabel to the plot

```python
np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))
```

Scatter plots are used to represent values for two different numeric variables and is a visual representation of how two variables relate to each other.

We can use scatter plots to explore the relationship between two variables, for example by looking for any correlation between them.

```python
plt.scatter(x,y,c='red',marker='x')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter Plot:X vs Y')
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

```python
import numpy as np

#line1: co-ord(0,0)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)                             #simplest method to generate subplots here: subplot(2,2,1) refers to a 2*2 figure subplot and third arg >1 refers to index for a figure
plt.plot(x,y)
plt.title('Line 1')

#line2: co-ord(0,1)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('Line 2')

#line3: co-ord(1,0)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('Line 3')

#line4: co-ord(1,1)
x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('Line 4')
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

--

A bar chart shows the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values (numerical) corresponding to those categories.




```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit,color='cyan')
plt.title('Bar Chart')
```

```python
plt.barh(company,profit,color='Blue')
plt.title('Horizontal Bar Chart')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

--


Box plot provides a graphical summary of a data set along with attributes about dataset like minimum, first quartile, median, third quartile, and maximum

```python
box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```

*   The markers at the extreme ('o') represent the outliers in the dataset.

*   **"Whisker Plot" is another name for the "Box Plot."** - The whiskers are the two lines that run from the minimum to the lower quartile (the box's beginning) and then from the higher quartile (the box's end) to the maximum.

*  The lower line in the box is 25th percentile, while to one at the top of the box is 75th percentile

*  The orange line represents the median, and shows how skewed the dataset is depending on wher eit is placed in the box (towards bottom or top)
